```python
filtro8 = df[(df['MONTO ADEUDADO'] < 6000) & (df['FECHA_FACTURA'] == '2023-03-14')]
filtro8
```

```python
filtro10 = df[~(df['MONTO ADEUDADO'] >= 6000) & ~(df['FECHA_FACTURA'] == '2023-03-14')]
filtro10
```
